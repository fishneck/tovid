```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

raw_data=pd.read_csv('../data/litcovid_words.csv')

raw_data = raw_data[raw_data.abstract.map(lambda x: len(str(x)))>50]
raw_data = raw_data.sort_values(by='month')
raw_data = raw_data[pd.notnull(raw_data.abstract)]
```

```python
raw_data.reset_index(drop=True,inplace=True)
raw_data
```

```python
def bert_format(s):
    s.replace('(i)','')
    s.replace('(ii)','')
    s.replace('(iii)','')
    s.replace('(iv)','')
    s.replace('(v)','')
    s.replace('(vi)','')
    s.replace('(',',')
    s.replace(')',',')
    return s 
    
```

```python
import math


def extract_motion_fea(batch,raw): 
    for i in range(math.ceil(raw.shape[0]/batch)):
        print('------------begin extracting features from the '+str(batch*i)+ ' with a batchsize of ' + str(batch))
        x = list(raw[batch*i:min(batch*(i+1),raw.shape[0])].abstract.map(lambda x: bert_format(x)).values)
        embedd = bc.encode(x)
        df = pd.DataFrame(embedd)
        print(df)
        doc_info = raw.drop('abstract',axis=1)[batch*i:min(batch*(i+1),raw.shape[0])]
        doc_info.reset_index(drop=True,inplace=True)
        result = pd.concat([doc_info,df],axis=1,ignore_index=True)
        result.rename({0:'index',1: 'title', 2:'doi', 3:'journal',4:'month'}, axis='columns',inplace=True)
            
        if i==0:
            print(result.shape)
            result.to_csv('../data/bert_embedd.csv',index=False)
        else:
            print(result.shape)
            result.to_csv('../data/bert_embedd.csv', mode='a', header=False,index=False)


        
extract_motion_fea(5,raw_data[:15])



```
